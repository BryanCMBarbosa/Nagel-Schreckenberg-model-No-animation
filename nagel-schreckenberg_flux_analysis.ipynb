```python
from nagel_schreckenberg_model import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
road_size = 200
num_cars_range = np.arange(1, 200)
max_speed = 5
brake_prob = 0.1
iterations = 1000
simulations_per_num_cars = 50
flux_by_num_car = {}
```

```python
for nc in num_cars_range:
    flux_by_num_car[nc] = np.array([])
    for i in range(simulations_per_num_cars):
        r = Road(road_size, nc, max_speed, brake_prob, iterations)
        r.run(terminal_printing=False)
        flux_by_num_car[nc] = np.append(flux_by_num_car[nc], r.calculate_average_flux())
```

```python
for nc in num_cars_range:
    for i in range(simulations_per_num_cars):
        plt.scatter(nc, flux_by_num_car[nc][i], alpha=0.05, c="blue")
plt.show()
```
